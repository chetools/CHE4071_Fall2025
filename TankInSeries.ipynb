```python
import numpy as np
from plotly.subplots import make_subplots
from scipy.integrate import solve_ivp
```

```python
A = 1.2 #m2
Aout = 0.2 #m2
g=9.81 #m/s2
qin0 = 0.5
h0 = ((qin0/Aout)**2)/(2*g)


def rhs(t, v):
    h=v
    qin = qin0 + 0.1*qin0*np.sin(25*t)
    return (qin - Aout*np.sqrt(2*g*h))
```

```python
tend= 20.
res=solve_ivp(rhs, (0, tend), [h0], dense_output=True, method='Radau')
res
```

```python
tplot = np.linspace(0, tend, 1000)
h1=res.sol(tplot)[0]
fig=make_subplots()
fig.add_scatter(x=tplot, y=h1, mode='lines', name='h1')
fig.update_layout(width=600, height=400, template='plotly_dark')
```
