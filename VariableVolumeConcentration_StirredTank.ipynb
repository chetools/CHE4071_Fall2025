```python
from importnb import Notebook
with Notebook():
    from che5 import sim, pid, TF1, TF2, shift

import numpy as np
import scipy as sp
from plotly.subplots import make_subplots
import plotly.io as pio
pio.templates.default = "plotly_dark"
import jax
import jax.numpy as jnp
jax.config.update("jax_enable_x64", True)

from plotly.subplots import make_subplots

from sympy.abc import s
from sympy import exp, Symbol, simplify
import scipy as sp
import scipy.signal as sig
```

```python
g = 9.81 #m/s2
A = 1.2 #m2
Aout = 0.2 #m2
qin_ss = 0.5 #m3/s
Cain_ss = 1e3 #mol/m3 = 1 M
h_ss = ((qin_ss/Aout)**2)/(2*g)
tend = 10 #s
```

```python
def rhs(t, vec):
    h, Ca = vec
    qin = qin_ss*1.1
    Cain = Cain_ss*0.95
    qout = Aout*np.sqrt(2*g*h)
    dh = (qin - qout)/A
    B = (qin*Cain - qout*Ca)/A
    dCa = (B - Ca*dh)/h
    return np.array([dh, dCa])

```

```python
res=sp.integrate.solve_ivp(rhs, (0,tend), [h_ss, Cain_ss], method='Radau', dense_output=True)
```

```python
tplot = np.linspace(0,tend,200)
h, Ca = res.sol(tplot)
fig=make_subplots(rows=1,cols=2)
fig.add_scatter(x=tplot, y=h, mode='lines', name='h', row=1, col=1)
fig.add_scatter(x=tplot, y=Ca, mode='lines', name='Ca', row=1, col=2)
fig.update_layout(width=800, height=400)
```
