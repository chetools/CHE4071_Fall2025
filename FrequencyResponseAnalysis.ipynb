```python
tau = 2.
K = 1.
Kc = 10/1.23/2
G = Kc*K/((tau*s+1))**3
```

```python
Gcl = G/(1+G)
```

```python
fig, mag, phase = bode(G)
fig.show()
```

```python
tplot, hplot = sim(Gcl, lambda t: 1.,N=1000,dt=0.1)
fig=make_subplots()
fig.add_scatter(x=tplot, y=hplot, mode='lines')
fig.update_layout(width=600,height=400)
```

```python
K = 1.
tau = 2.
Kc = 2
taui = 1e6
taud = 3.  #dead-time
Gc = Kc*(1+1/(taui*s))
Gp = K/(tau*s+1)  * exp(-taud*s)
Gcl = Gc*Gp/(1+Gc*Gp)
```

```python
t, y = sim(Gcl, lambda t: 1., N=1000, dt=0.1)
fig=make_subplots()
fig.add_scatter(x=t, y=y, mode='lines')
fig.update_layout(width=600,height=400)
```
