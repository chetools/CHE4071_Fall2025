```python
def sumsqerr(v):
    K, tau, theta = v
    ymodel = np.where(t<theta, 0., K*(1-np.exp(-(t-theta)/tau)))
    return np.sum((y-ymodel)**2)
```

```python
K_fit, tau_fit, theta_fit = sp.optimize.minimize(sumsqerr, [K,tau,theta]).x
```

```python
fig=make_subplots()
fig.add_scatter(x=t, y=y, mode='markers')
fig.add_scatter(x=t, y=np.where(t<theta_fit, 0., K_fit*(1-np.exp(-(t-theta_fit)/tau_fit))), mode='lines')
fig.update_layout(width=600,height=400)
```

```python
#Smith and Corripio PI parameters
PA, PB = 0.586, -0.916
IA, IB = 1.03, -0.165
Y = PA*(theta_fit/tau_fit)**PB
Kc = Y/K_fit
taui = tau_fit/(IA + IB*theta_fit/tau_fit)
Gc = Kc*(1 + 1/(taui*s))
```

```python
Gcl = Gc*Gp/(1+Gc*Gp)
t, y = sim(Gcl, lambda t: 1., N=1000, dt=0.1)
fig=make_subplots()
fig.add_scatter(x=t, y=y, mode='lines')
fig.update_layout(width=600,height=400)
```
